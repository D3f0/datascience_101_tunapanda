```python
# Small script that downloads from github and unzips
from pathlib import Path
import requests
import tempfile
import sh

if not Path('.').joinpath('test_db-master').is_dir():
    with tempfile.NamedTemporaryFile("w+b", suffix='test_data.zip') as tmp:
        print(f"Downloading to {tmp.name}")
        with requests.get('https://github.com/datacharmer/test_db/archive/master.zip') as rq:
            tmp.write(rq.content)
        print("Creating directory for test database")
        print(f"Unzipping {tmp.name}")
        !unzip {tmp.name}
    
```

```python
# Delete container
!docker stop mysql | xargs docker rm
```

```python
test_db_dir = (Path('.') / 'test_db-master').absolute()
!docker run -d -p 3306:3306 -v {test_db_dir}:/test_db --name mysql -e MYSQL_ROOT_PASSWORD=password mysql
```

```python
# Still running?
!docker ps | grep mysql
```

```python
# Has the test data?
!docker exec mysql ls /test_db/employees.sql
```

```python
%%writefile test_db-master/load.sh
#!/bin/bash
cd /test_db/
while : ; do
    mysql --password=password < employees.sql && exit 0 ;
    sleep 1 ;
done
```

```python
!docker exec -ti mysql bash /test_db/load.sh
```

# Connecting and reading data <a id='reading'></a>

```python
import pandas as pd

try:
    import pymysql
except ImportError:
    # There are many SQL conectors, Python does not come with one by default
    !pip install PyMySQL


from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:password@localhost/employees')
```

```python
pd.read_sql_table('employees',con=engine)
```

```python
df = pd.read_sql('select * from employees LIMIT 100',con=engine)
df
```
