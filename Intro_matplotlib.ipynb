### Types of plots

**Line Plot**: a type of plot which displays information as a series of data points called markers connected by straight lines. In this type of plot, we need the measurement points to be **ordered** (typically by their x-axis values). This type of plot is often used to visualize a trend in data over intervals of time.

```python
years = [1990, 1995, 2000, 2005, 2010, 2015, 2016, 2017, 2018]
total_populations_kenya = [23402507, 27346456, 31450483, 36048288,
                           41350152, 47236259, 48461567, 49699862,
                           50950879]

plt.plot(years, total_populations_kenya, c='m')
plt.title("Year vs Population in Kenya")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.show()
```

```python
%matplotlib inline

years = [1990, 1995, 2000, 2005, 2010, 2015, 2016, 2017, 2018]
total_populations_arg = [32729739, 34994814, 37057452, 39145488, 
                         41223889, 43417765, 43847430, 44271041,
                         44688864]

plt.plot(years, total_populations_kenya, c='m', label='Kenya')
plt.plot(years, total_populations_arg, c='b', label='Argentina')
plt.title("Year vs Population in Argentina and Kenya")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.legend()
```

**Scatter plot**: this type of plot shows all individual data points. Here, they aren’t connected with lines. Each data point has the value of the x-axis value and the value from the y-axis values. This type of plot can be used to display **trends** or **correlations**. In data science, it shows how 2 variables compare.

```python
temp = [30, 32, 33, 28.5, 35, 29, 29]
ice_creams_count = [100, 115, 115, 75, 125, 79, 89]

plt.scatter(temp, ice_creams_count)
plt.title("Temperature vs. Sold ice creams")
plt.xlabel("Temperature")
plt.ylabel("Sold ice creams count")
```

**Histogram**: an accurate representation of the distribution of numeric data. 
To create a histogram:
- we divide the entire range of values into a series of intervals.
- we count how many values fall into each interval. The intervals are also called bins. 

The bins are consecutive and non-overlapping intervals of a variable. They must be adjacent and are often of equal size.

```python
numbers = [0.1, 0.5, 1, 1.5, 2, 4, 5.5, 6, 8, 9]

plt.hist(numbers, bins = 3)
plt.xlabel("Number")
plt.ylabel("Frequency")
```

**Box plot**: a way to show the distribution of values based on the five-number summary: minimum, first quartile, median, third quartile, and maximum.
- The minimum and the maximum are just the min and max values from our data. 
- The median is the value that separates the higher half of a data from the lower half. ($1, 2, 5, 6, 8, 9$, your median will be $(5 + 6) / 2 = 5,5)$.
- The first quartile is the median of the data values to the left of the median in our ordered values. For example, if we have the numbers $1, 3, 4, 7, 8, 8, 9$, the first quartile is the median from the $1, 3, 4$ values, so it’s $3$.
- The third quartile is the median of the data values to the right of the median in our ordered values. For example, if we use these numbers $1, 3, 4, 7, 8, 8, 9$ again, the third quartile is the median from the $8, 8, 9$ values, so it’s $8$.
- This type of plot can also show outliers.

```python
values = [1, 2, 5, 6, 6, 7, 7, 8, 8, 8, 9, 10, 21]

plt.boxplot(values)
plt.yticks(range(1, 22))
plt.ylabel("Value")
```

**Bar chart**: represents categorical data with rectangular bars. Each bar has a height corresponds to the value it represents. It’s useful when we want to compare a given numeric value on different categories.

```python
labels = ["JavaScript", "Java", "Python", "C#"]
usage = [69.8, 45.3, 38.8, 34.4]

# Generating the y positions. Later, we'll use them to replace them with labels.
y_positions = range(len(labels))

# Creating our bar plot
plt.bar(y_positions, usage, c=)
plt.xticks(y_positions, labels)
plt.ylabel("Usage (%)")
plt.title("Programming language usage")
```

**Pie chart**: a circular plot, divided into slices to show numerical proportion. They are widely used in the business world. However, many experts recommend to **avoid them.** The main reason is that it’s difficult to compare the sections of a given pie chart.

## Now lets plot from pandas!

```python
import pandas as pd
import numpy as np
```

```python
df4 = pd.DataFrame({'a': np.random.randn(1000) + 1, 'b': np.random.randn(1000),
                    'c': np.random.randn(1000) - 1})
df4.plot.hist(alpha=0.5)
```

```python
df4.plot.hist(stacked=True, bins=20)
```

```python
df4.plot.box()
```

```python
color = {'boxes': 'DarkGreen', 'whiskers': 'DarkOrange',
         'medians': 'DarkBlue', 'caps': 'Gray'}
df4.plot.box(color=color)
```

```python
s = pd.Series([2, 3, 5, -1, 0])
s, s.cumsum()
```

```python
plt.title('cumsum of S')
s.cumsum().plot()
plt.xlabel('position of element')
plt.ylabel('cumsum values')
```

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot.area(stacked=False);
```

```python
df.plot.area()
```
